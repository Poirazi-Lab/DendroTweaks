```python
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
import dendrotweaks as dd
```

```python
import os
path_to_data = os.getcwd()
path_to_data
```

```python
path_to_model = os.path.join(path_to_data, "Park_2019")
Park_model = dd.Model(path_to_model)
# Park_model.verbose = True
Park_model.load_morphology('main')
Park_mechs = set(Park_model.list_mechanisms())
Park_model.load_biophys('main')
Park_model.load_stimuli('current_soma_depol')

fig, ax = plt.subplots(1, 1, figsize=(10, 4))
Park_model.simulator.run(1000);
Park_model.simulator.plot_voltage(ax=ax)
```

```python
path_to_model = os.path.join(path_to_data, "Poirazi_2003")
Poirazi_model = dd.Model(path_to_model)
# Poirazi_model.verbose = True
Poirazi_model.load_morphology('main')
Poirazi_mechs = set(Poirazi_model.list_mechanisms())
Poirazi_model.load_biophys('main')
Poirazi_model.load_stimuli('current_soma_depol')

fig, ax = plt.subplots(1, 1, figsize=(10, 4))
Poirazi_model.simulator.run(900);
Poirazi_model.simulator.plot_voltage(ax=ax)
```

```python
path_to_model = os.path.join(path_to_data, "Hay_2011")
Hay_model = dd.Model(path_to_model)
# Hay_model.verbose = True
Hay_model.load_morphology('main')
Hay_mechs = set(Hay_model.list_mechanisms())
Hay_model.load_biophys('main')
Hay_model.load_stimuli('current_soma_depol')

fig, ax = plt.subplots(1, 1, figsize=(10, 4))
Hay_model.simulator.run(1000);
Hay_model.simulator.plot_voltage(ax=ax)
```
